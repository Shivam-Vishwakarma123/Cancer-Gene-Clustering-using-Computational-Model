```python
import numpy as np
import matplotlib.pyplot as plt
# Dataset generation
data_dict = {'K-Means':0.170, 'Hierarchical Clustering':0.172, 'DBSCAN':0.166, 'SOM':0.134}
algo = list(data_dict.keys())
values = list(data_dict.values())
fig = plt.figure(figsize = (7, 5))
#  Bar plot
plt.bar(algo, values, color ='blue',
        width = 0.25)
plt.xlabel("Clustering algorithm")
plt.ylabel("Silhouette Index Values")
plt.title("Silhouette Index Value")
plt.show()
```
